#CAPITULO 2 Variables:

```python
message = "Hello world!"
print(message)
```

El mensaje que salió de la consola es similar al anterior, pero la estructura es distinta. se creó algo llamado *variable* cada variable guarda un *valor*

```python
message = "Hello World!"
print(message)

message = "Hello Python Crash Course world!"
print(message)
```

A pesar de que es la misma variable, esta cambió de nombre por lo que el mensaje original "Hello World!" fue reemplazado por la segunda modificación. Cuando se usan variables en python hay algunas reglas, las cuales al entenderlas evitamos tener errores.
1. Los nombres de las varialbes pueden contener **solamente** letras, numberos o guiónes bajos (underscore), estos pueden empezar con letras o guíon bajo, pero nunca con un numbero, por ejemplo puede una variable llamarse message_1 pero no 1_message.
2. Los espacios no son permitidos en los nombres de la variable, pero se pueden usar los guiones bajos para separar las variables.
3. Se debe evitar usar como variables palabras reservadas por python, estas son reservadas para uso directo en el funcionamiento de python, por ejemplo "if"
4. nos nombres de las variables deben ser cortos, pero *descriptivos*, por ejemplo es mejor *name* que *n*, *student_name* que *s_n*
5. debemos tener cuidado al utilizar la minuscula l o la mayuscula O pues pueden ser confundidas con 1 y 0.

Es una buena practica aprender a nombrar nuestras variables especialmente cuando los programas pueden ser grandes y complicados. a medida que se escriben codigos o se leen la de otras personas, se entiende más la necesidad de poner nombres significativos.

se recomienda nombrar a toda variable en minusculas, no hay problema en nombrarla en mayuscula, pero para evitar problemas al momento de implementarlo es mejor hacerlo en minuscula.

##Cambiar mayusculas o minusculas
Uno de las tareas más simples que podemos hacer es cambiar las mayusculas dentro de las palabras en una cadena:

```python
name ="ada lovelace"
print(name.title())
```

En este ejemplo, se convirtió de minusculas a mayusculas. uno de los metodos utilizados aquí fue ***.title()*** esto permite convertir cáda variable en mayusculas. cada metodo debe ser unido a la variable con un punto y seguido por un par de parentesis si es que la función no necesita argumentos. el mismo ejemplo ocurre con la funcion ***.lower()***

las variables se pueden combinar o concatenar, por ejemplo para unir nombres completos.

```python
first_name = "ADA"
last_name = "lovelace"
full_name = first_name + " " + last_name

print(full_name.title())
```

Python usa el simblolo suma o plus (+) para combinar las cadenas. Ese metodo se llama concatenación, se puede usar la concatenación para mostrar mensajes completos utilizando información almacenada en variables.

```python
first_name = "ADA"
last_name = "lovelace"
full_name = first_name + " " + last_name

print("Hola, " + full_name.title() +"!")
```

El almacenar variables que contienen el contenido de otra es util para poder cumplir con funciones distintas.
