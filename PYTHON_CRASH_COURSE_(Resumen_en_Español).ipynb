## AGREGAR ESPACIOS EN BLANCO

Los espacios en blanco hacen referencia a los caracteres que no se imprimen como espacios, tabulaciones y finales de linea.
\t sirve para tabular
\n sirve para agregar saltos de linea.

```python
print("Python")
print("\tPython")
print("Languages:\nPython\nC\nJavaScript")
print("Languages:\n\tPython\n\tC\n\tJavaScript")
```

Para un programador "python" y "python " se miran igual, pero para un programa son dos cadenas completamente distintas, pues python detecta ese espacio extra y este es significane a menos que le informemos al programa que no.
Es importante pensar en los espacios en blanco, por que, en ocaciones quisieramos comparar el contenido de dos cadenas.
Python puede revisar los espacios en blanco extra al lado derecho o iquierdo de las cadenas, para hacerlo se puede usar el metodo ***.rstrip()*** o ***.lstrip()***


```python
favorite_language = "python "
print(favorite_language)
print(favorite_language.rstrip())
```

Este espacio es desaparecido por un tiempo, sin embargo al invocar nuevamente la función sin este metodo esta vuelve a tener los espacios, la recomendación es almacenar la función dentro de otra donde solamente se recupere la función corregida.

## NÚMEROS

Los numeros son usados dentro del código para poder mantener punteos en los juegos por ejemplo, o representar datos en las visualizaciones, almacenar informacion en aplicaciones web.
Python trata los numeros de maneras distintas, dependiendo del uso que se le darán.
### ENTEROS
Los numeros enteros pueden ser sumados (+) , restados (-), multiplicados (*) y divididos (/).
en caso de exponentes (**) usamos doble multiplicación.
Python respeta tambien el orden de las operaciones, podemos utilizar parentesis para modificar el orden de operaciones para poder evaluar en el orden que especifiquemos.
eje:


*   2+3 = 5
*   3-2 = 1
* 2*3 = 6
* 3/2 = 1.5


### FLOAT (FRACCIONES)
Python llama *float* a cada numero que tiene un número decimal.
* 3**2 = 9
* 3**3 = 27
* 10**6 = 1000000

Sin embargo debemos tener cuidado porque de podríamos tener un numero decimal arbitrario en nuestras respuestas.
* 0.2+0.1 = 0.30000000000000004
esto pasa en todos los lenguages, sin embargo python trata de encontrar los resultados de la manera más precisa porsible. en ocaciones es un poco dificil como otras computadoras interpretan o representan esos números. En los proyectos se estudiará más a fondo el tema de los decimales para evitar problemas.

### EVITAR Type Error (Errores de tipo)
A menudo vamos a queres usar valores de variables dentro de un mensage, por ejemplo si queremos desear feliz cumpleaños a alguien podríamos hacerlo de la siguiente manera:




```python
age = 23
message = "Happy " + age + "rd Birthday!"
print(message)
```

Ese código da como resultado un error *TypeError*, lo cual significa que Python no puede reconocer el tipo de infomrración que estamos utilizand. En este ejemplo está identificando dos tipos distintos según la naturaleza de la cadena, una cadena es tipo texto (message) y la otra tipo número (age). para solucionarlo es muy sencillo, tenemos que convertir el número en texto, para ello podemos utilizar la función ***.str()*** la cual representa los valores como un texto

```python
age = 23
message = "Happy " + str(age) + "rd Birthday!"
print(message)
```
