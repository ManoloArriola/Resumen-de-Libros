## AGREGAR ESPACIOS EN BLANCO

Los espacios en blanco hacen referencia a los caracteres que no se imprimen como espacios, tabulaciones y finales de linea.
\t sirve para tabular
\n sirve para agregar saltos de linea.

```python
print("Python")
print("\tPython")
print("Languages:\nPython\nC\nJavaScript")
print("Languages:\n\tPython\n\tC\n\tJavaScript")
```

Para un programador "python" y "python " se miran igual, pero para un programa son dos cadenas completamente distintas, pues python detecta ese espacio extra y este es significane a menos que le informemos al programa que no.
Es importante pensar en los espacios en blanco, por que, en ocaciones quisieramos comparar el contenido de dos cadenas.
Python puede revisar los espacios en blanco extra al lado derecho o iquierdo de las cadenas, para hacerlo se puede usar el metodo ***.rstrip()*** o ***.lstrip()***


```python
favorite_language = "python "
print(favorite_language)
print(favorite_language.rstrip())
```

Este espacio es desaparecido por un tiempo, sin embargo al invocar nuevamente la función sin este metodo esta vuelve a tener los espacios, la recomendación es almacenar la función dentro de otra donde solamente se recupere la función corregida.

# CAPITULO 3 INTRODUCCIÓN A LAS LISTAS

En este y el próximo capítulo trabajaremos con elementos en lista. Las listas permiten almacenar información en un lugar, ya sea que tengas pocos items o un millon de ellos. Las listas dentro de Python son una de las funcionalidades más poderosas para que los nuevos programadores aprendan los conceptos más importantes.

**Qué es una lista?**
Una lista es una colección de items en un orden particular. podemos hacer listas que incluyan las letras del abecedario, los digitos de 0 - al 9, nombres de personas, etc.
Podemos agregar todo lo que deseemos en esa lista, y estos no tienen que estar obligatoriamente relacionados de una manera particular. Debido a que los elementos de la lista usualmente contienen más de un item. es una buenda idea nombrar nuestras listas en plural.

En python, los corchetes cuadrados **[ ]** son los que indican que se trata de una lista, los elementos dentro de la lista están separados por comas (,)

```python
bicycles = ["trek","cannondale","redline","specialized"]
print(bicycles)
```

Cuando pedimos a Python que imprima una lista, regresará la representación de la misma, incluyendo los corchetes cuadrados.
Debido a que esto no es lo que los usuarios quieren ver, vamos a aprender a acceder a los elementos de manera individual.

## Accediendo a los elementos de la lista

Las listas son colecciones ordenadas, y podemos acceder a cualquier elemento de la lista al decirle a python la posición o indice del item deseado. Para acceder a elementos dentro de la lista, escribimos el nombre de la lista, seguido por el indice del item encerrado en los corchetes cuadrados. por ejemplo con la lista bicicletas (bicycles) es de esta manera:

```python
bicycles = ["trek","cannondale","redline","specialized"]
print(bicycles[0])
```

Cuando pedimos un item especifico en una lista Python devuelve unicamente el item solicitado.
a este resultado tambienn se pueden adicionar formatos como se aprendió en el capitulo 2

```python
bicycles = ["trek","cannondale","redline","specialized"]
print(bicycles[2].title())
```

##Las posiciones inician en 0 no en 1
Python considera como el primer elemento en una lista con el # 0, no la posición 1. Esto es verdad para la mayoría de lenguajes de programación, y la razón para hacer con ella que esté implementada a un nivel bajo. Si estamos recibiendo un resultado que no es el correcto, debemos determinar si nuestro error es por el calculo de uno.

El segundo item en una lista tiene en indice 1. Usando este simple sistema continuo, podemos obtener cualquier objeto al restar un elemento del indice.

Pero que pasa cuando queremos acceder al ultimo de los items en la lista, y no conocemos la longitud o cantidad de elementos. Bueno, para ello python tiene un metodo sencillo, este es usar el indice ***-1***. Con este sencillo metodo, pithon regresa el ultimo de la lista aún cuando no sabemos cuantos existen dentro.

```python
bicycles = ["trek","cannondale","redline","specialized"]
print(bicycles[-1])
```

```python
bicycles = ["trek","cannondale","redline","specialized"]
message = "My first bicycle was a "+ bicycles[0].title() + "."

print(message)
```

##Modificar los elementos de la lista.
La sintaxis para modificar los elementos de la lista es igual a la de acceder a los elementos de ella. Para cambiar el elemento de la lista, usamos el nombre de la lista, seguido por el indice del elemento que queremos cambiar y luego proveer un nuevo valor que queremos que tenga.
Por ejemplo, digamos que tenemos un listado de motocicletas, y que el primer item en nuestra lista es "honda".
Cómo podríamos cambiar el valor del primer elemento?

```python
motorcycles = ["honda","yamaha","suzuki"] #El codigo aqui define a la lista original con honda como su primer elemento
print(motorcycles)
motorcycles[0] = "ducati"  # Este código cambia el valor del elemento indexado en la primera posición, dejando el resto de la lista de la misma manera
print(motorcycles)
```

##Agregando elementos a la lista.

Podríamos querer agregar nuevos elementos por muchas razones, por ejemplo queremos hacer que aparezcan nuevos aliens en el juego, agregar más datos a nuestra visualización, o agregar nuevos usuarios registrados al sitio web que construimos. Python provee distintas maneras de agregar datos a nuestras listas existentes:

###Adjuntando elementos al final de la lista.

La forma más simple de agregar elementos nuevos a nuestra lista es por medio de *.append()* cuando los adjuntamos estos son enviados al final de la lista. usaremos el mismo ejemplo de las motocicletas.

```python
motorcycles = ["honda","yamaha","suzuki"]
print(motorcycles)

motorcycles.append("ducati")
print(motorcycles)
```

El metodo .append() agrega al final de la lista sin afectar ningun otro de los elementos existentes. Este es uno de los metodos más facil para construir listas dinamicas. Por ejemplo podríamos empezar con un listado bacío y luego ir agregando items a nuestra lista usando declaraciones con append.

```python
motorcycles =[]

motorcycles.append("honda")
print(motorcycles)

motorcycles.append("yamaha")
print(motorcycles)

motorcycles.append("suzuki")
print(motorcycles)
```

Construir listas de esta manera es muy comun, porque a menudo, no sabemos los datos de nuestros usuarios que queremos almacenar en un programa hasta que el programa está corriendo. Les damos el control a nuestros usuarios cuando definimos una lista vacía y esperamos que los usuarios agreguen los valores. cada adjunción (append) brinda un valor a la lista que se ha creado.

###Insertar elementos en la lista.

Podemos agregar elementos nuevos a una lista en cualquier posición que deseemos por medio del metodo *.insert()*. Lo hacemos al especificar el indice del nuevo elemento y el valor que tendrá este.

```python
motorcycles = ["honda","yamaha","suzuki"]

motorcycles.insert(1,"ducati")
print(motorcycles)
```

En este ejemplo, el codigo insertó el valor "ducati" en la posición 1 (recordemos que el incide empieza a contar desde 0). Este metodo abre un espacio en la posición deseada y almacena el valor en esta ubicación. Esta operación cambia la posición de todos los items posicionados a la derecha.

###Remover elementos de una lista.

A menudo, vamos a querer remover los items o un conjunto de items de una lista. Por ejemplo el jugador dispara y baja a los aliens del espacio. Es probable que quieras removerlos de la lista de Aliens activos, o cuando un usuario decide cancelar su cuenta de la aplicación web que creaste, vas a querer remover a ese usuario del listado de usuarios activos. Puedes remover un item de acuerdo a su posición en la lista o de acuerdo a su valor.

####Remover un item usando del Statement

Si saves la posicion del item que quieres remover, puedes usar la función *del*

```python
motorcycles = ['honda', 'ducati', 'yamaha', 'suzuki']
print(motorcycles)

del motorcycles[0]
print(motorcycles)
```

El código uso la función del para eliminar el primer item de la lista.
Podemos remover items de cualqueir posición de la lista usando la posición exacta donde sabemos que está almacenada.

```python
motorcycles = ['honda', 'ducati', 'yamaha', 'suzuki']
print(motorcycles)

del motorcycles[2]
print(motorcycles)
```

En este ejemplo eliminó a yamaha de la lista original por tener la posición 3 (Indice 2) de la lista.

En ambos ejemplos, no podremos acceder más al valor removido de este listado.

####Remover por el metodo pop()

Algunas veces querremos usar el valor del item despues de removerlo. por ejemplo queremos obtener la posición x & y  de un alien que ha sido eliminado, para poder dibujar una exploción en esa posición. en una aplicación web, querriamos remover el usuario de la lsita de usuarios activos y trasladarlo a la lista de usuarios inactivos.

El metodo *pop()* remueve el ultimo elemento de la lsita, pero te deja trabajar con el despues de removerlo. El término "pop" proviene de pensar en una lista como una pila de elementos y de extraer un elemento de la parte superior de la pila.

```python
motorcycles = ['honda', 'ducati', 'yamaha', 'suzuki']
print(motorcycles)

popped_motorcycle = motorcycles.pop()
print(motorcycles)
print(popped_motorcycle)
```

comenzamos definiendo los elementos de la tabla, al hacer pop guardamos este dato quitado de la tabla original, pero al ser relacionado a una nueva variable, este queda almacenado allí. pudiendo utilizarlo para hacer referencia al dato que fue removido. Esto es util para poder manejar informacion con referencia a lo que se está trasladando de una tabla a otra.

```python
motorcycles = ['honda', 'ducati', 'yamaha', 'suzuki']
print(motorcycles)

last_owned = motorcycles.pop()
print("La ultima motocicleta que compré fue una " + last_owned.title() + ".")
print(motorcycles)
```

Podemos hacer el pop de los items de cualquier posición según necesitemos. mencionamos que pop() remueve el ultimo elemento y esto será así siempre que no brindemos una posición especifica para realizarlo.

```python
motorcycles = ['honda', 'ducati', 'yamaha', 'suzuki']
print(motorcycles)

last_owned = motorcycles.pop(1)
print("La ultima motocicleta que compré fue una " + last_owned.title() + ".")
print(motorcycles)
```

En esete ejemplo escogimos la posición 1 por lo que "ducati" fue removida de la lista.



*   Recuerda que cada ve que uses pop, el item será removido de la lista.
*   Cuando querramos eliminar un item de una lista y que este no sea usado de ninguna manera, utilizamos el metodo *del*


####Remover por valor del item

Algunas veces no vamos a poder saer la posición del valor que queremos remover de la lista, y solamente conocemos el valor del item que queremos quitar, para ello podemos usar el metodo *remove()*
veamos el ejemplo

```python
motorcycles = ['honda', 'ducati', 'yamaha', 'suzuki']
print(motorcycles)
for motorcycle in motorcycles:
  motorcycle == motorcycles.pop(-1)
  print("La ultima motocicleta que compré fue una " + motorcycle.title() + ".")
  print(motorcycles)
```

##Organizando una Lista.

A menudo las listas se crearan en un orden impredecible, por lo que no siempre tendrás el control que los usuarios darán sus datos. Aunque esto es inevitable en la mayoría de los casos, frecuentemente querras presentar tu información en un orden particular. Algunas veces querras preservar el orden original de tu lista y otras veces querras cambiar el orden origianl. Python te da varias maneras de ordenar tu lista, dependiendo de la situación.

###Ordenar una lista permanente con el metodo sort()

El metodo *sort()* hace relativamente facil ordenar una lista. Imagina que tienes un lsitado de carros y quieres cambiar el orden de la lista y guardarlo alfabeticamente. Para guardar la tarea simple, asumamos que todos los valores estan en minusculas.


```python
cars = ['bmw','audi','toyota','subaru']
cars.sort()
print(cars)
```

El metodo sort() nos hace los cambios de la lista de manera permanente. los carros ahora están en orden alfabetico y estos no podrán ser revertidos al orden original. puedes tambien ordenar de una manera reversa alfabetica cambiando el argumento *reverse=True* por medio del metodo sort(). El siguiente ejemplo almacena de manera reversa

```python
cars = ['bmw','audi','toyota','subaru']
cars.sort(reverse=True)
print(cars)
```
